### 快排

```python
import random
nums = [3, 2, 1, 5, 6, 4, 7, 15, 9, 11, 10]
def partition(nums, left, right):
    pivot_idx = random.randint(left, right)
    nums[pivot_idx], nums[right] = nums[right], nums[pivot_idx]
    pivot = nums[right]
    i = left-1
    for j in range(left, right):
        if nums[j] <= pivot:
            i+=1
            nums[i], nums[j] = nums[j], nums[i]
    nums[i+1], nums[right] = nums[right], nums[i+1]
    return i+1
def quick_sort(nums, left, right):
    if left > right:
        return 
    pos = partition(nums, left, right)
    quick_sort(nums, left, pos-1)
    quick_sort(nums, pos+1, right)

quick_sort(nums, 0, len(nums)-1)
print(nums)
```

```python
import random
nums = [3, 2, 1, 5, 6, 4, 7, 15, 9, 11, 10]
def partition(nums, left, right):
    # 保持你的随机选择pivot的逻辑
    pivot_idx = random.randint(left, right)
    # 现在右边保存一下pivot
    nums[pivot_idx], nums[right] = nums[right], nums[pivot_idx]
    pivot = nums[right]

    i = left - 1
    for j in range(left, right):
        if nums[j] <= pivot:    # 如果j小于pivot，就要发动i++了，把原来放在i+1位置的大的数和小数交换！！！
            i+=1
            nums[i], nums[j] = nums[j], nums[i]     #保证小的都在pivot左侧
        
    nums[i+1], nums[right] = nums[right], nums[i+1]
    return i+1      # pivot所在位置

def quick_sort(nums, left, right):
    if left > right:
        return 
    pos = partition(nums, left, right)
    quick_sort(nums, left, pos - 1)
    quick_sort(nums, pos + 1, right)

quick_sort(nums, 0, len(nums)-1)
print(nums)
```

### 大顶堆用于升序排序

```python
def heapify(arr, n, i):
    largest = i
    left = 2*i+1
    right = 2*i+2
    if left < n and arr[largest] < arr[left]:
        largest = left
    if right < n and arr[largest] < arr[right]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
arr = [12, 11, 15, 17, 13, 10, 6, 5, 8, 7, 9, 4, 3, 2, 1] 
n = len(arr)
for i in range(n//2, -1, -1):
    heapify(arr, n, i)
print('初始化的堆，只实现了父节点>子节点:',arr)
```

```python
for i in range(n-1, 0, -1):
    arr[i], arr[0] = arr[0], arr[i] 
    print('没用堆排序',arr)
    heapify(arr, i, 0)
    print('用了堆排序', arr)
```

### 小顶堆用于降序排序

```python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[i] > arr[left]:
        largest = left
    if right < n and arr[largest] > arr[right]:
        largest = right
    
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

arr = [12, 11, 15, 17, 13, 10, 6, 5, 8, 7, 9, 4, 3, 2, 1] 
n = len(arr)
for i in range(n//2, -1, -1):
    heapify(arr, n, i)

print('初始化的堆，只实现了父节点<子节点:',arr)

for i in range(n-1, 0, -1):
    arr[i], arr[0] = arr[0], arr[i] 
    print('没用堆排序',arr)
    heapify(arr, i, 0)
    print('用了堆排序', arr)
```

### 约瑟夫问题

```python
n, k = 15, 4
start = 0
arr = [i for i in range(0, n)]
def leave(arr, start, k):
    if len(arr) == 1:
        return arr
    tmp = arr.pop((start + k - 1) % len(arr))
    print(tmp, '被删除')
    return leave(arr, (start + k - 1) % len(arr), k)

leave_result = leave(arr, start, k)
print(leave_result)
```
