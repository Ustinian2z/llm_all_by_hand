```python
num = 10000
for i in range(0, 16, 2):
    even = num & (1 << i)
    print(even)
```

```python
n = 1000
while n != 0:
    tmp = n & 1
    n >>= 1
    print(tmp)
```

```python
line = '1' + 27 * '0' + "1101"
neg = 0xffffffff
num = int(line, 2)
print(num)
if line[0] == '1':
    num = (num - 1) ^ neg
    num = -num 
print(num)
```
