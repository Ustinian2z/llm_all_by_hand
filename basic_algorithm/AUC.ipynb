```python
def AUC(label, pre):
    pos_idx = [i for i in range(len(label)) if label[i] == 1]
    neg_idx = [i for i in range(len(label)) if label[i] == 0]
    auc = 0
    for idx in pos_idx:
        for jdx in neg_idx:
            if pre[idx] > pre[jdx]:
                auc += 1
            elif pre[idx] == pre[jdx]:
                auc += 0.5
    return auc / (len(pos_idx) * len(neg_idx))

if __name__ == '__main__':
    label = [1, 0, 0, 0, 1, 0, 1, 0]
    pre = [0.9, 0.8, 0.3, 0.1, 0.4, 0.9, 0.66, 0.7]
    print(AUC(label, pre))
```

```python
# nlogn复杂度做法
def AUC(label, pre):
    data = sorted(zip(pre, label), key=lambda x:x[0])
    pre_sorted, label_sorted = zip(*data)

    ranks = []
    for i, (pre, label) in enumerate(data):
        if label == 1:
            j = i+1
            while j < len(pre_sorted) and pre_sorted[j] == pre:
                j+=1
            j-=1
            avg_rank = (i+j)/2
            ranks.append(avg_rank+1)
    
    m = sum(label_sorted)
    n = len(label_sorted) - m
    sum_ranks = sum(ranks)
    auc = (sum_ranks-m*(m+1)/2) / (m*n)
    return auc

if __name__ == '__main__':
    label = [1, 0, 0, 0, 1, 0, 1, 0]
    pre = [0.9, 0.8, 0.3, 0.1, 0.4, 0.9, 0.66, 0.7]
    print(AUC(label, pre))
```
